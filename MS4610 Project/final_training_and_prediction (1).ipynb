```python
def ModeNan(dataset):
    for i in range(len(dataset.columns)):
        if(i<=14):
            dataset[dataset.columns[i]].fillna(dataset[dataset.columns[i]].mode(),inplace = True)
```

```python
def changeNan(dataset):
    for i in range(len(dataset.columns)):
        if(i>14):
            dataset[dataset.columns[i]].fillna(dataset[dataset.columns[i]].mean(),inplace = True)
        
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
xgc = XGBClassifier(learning_rate = 0.1,max_depth = 5,n_estimators = 600,subsample=0.8,colsample_bytree=0.8,seed=27)
```

```python
Y = TRAIN['target']
```

```python
X = TRAIN[TRAIN.columns.drop(list(TRAIN.filter(regex='target')))]
X= TRAIN[TRAIN.columns.drop(list(TRAIN.filter(regex='cust_id')))]
x_test = TEST[TEST.columns.drop(list(TEST.filter(regex='cust_id')))]
```

```python
X.drop(X.columns[0],axis=1,inplace=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 10)
```

```python
xgc.fit(x_train,y_train)
```

```python
y_pred = xgc.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print (classification_report(y_test,y_pred))
```
